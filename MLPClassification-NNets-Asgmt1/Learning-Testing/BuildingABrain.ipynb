```python
import tensorflow as tf
tf.config.list_physical_devices('GPU')
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (valid_images, valid_labels) = fashion_mnist.load_data()
```

```python
import matplotlib.pyplot as plt 
# The question number to study with. 

for i in range(1,100):
    plt.figure()
    plt.imshow(train_images[i],cmap='gray')
    plt.colorbar()
    plt.grid(False)
    plt.show()
    

```

```python
data_idx = 45
valid_images[data_idx]
```

```python
number_of_classes = train_labels.max() + 1
number_of_classes
```

```python
model = tf.keras.Sequential([
tf.keras.layers.Flatten(input_shape=(28, 28)),
tf.keras.layers.Dense(number_of_classes)])
model
```

```python
model.compile(optimizer='adam',
loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
metrics=['accuracy'])
```

```python
history = model.fit(train_images, train_labels, epochs=5, verbose=True, validation_data=(valid_images,valid_labels))
```

```python
model.predict(train_images[0:10])
```

```python
data_idx = 45 # The question number to study with. Feel free to change up to 59999.
plt.figure()
plt.imshow(train_images[data_idx], cmap='gray')
plt.colorbar()
plt.grid(False)
plt.show()
x_values = range(number_of_classes)
plt.figure()
plt.bar(x_values, model.predict(train_images[data_idx:data_idx+1]).flatten())
plt.xticks(range(10))
plt.show()
print("correct answer:", train_labels[data_idx])
```
